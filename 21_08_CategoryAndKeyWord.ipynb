Laptop categorisi icin ANAHTAR KELIMELERI bul

```python
def rate_word(comment,dictionary):
    splitComment = comment.split(" ")
    for eachcomment in splitComment:
        eachcomment = eachcomment.lower()
        if eachcomment in dictionary:
            x = dictionary[eachcomment]
            dictionary[eachcomment] = x+1
        else:
            dictionary[eachcomment] = 1
```

```python
dictionary_laptop = {}
allcomment = df_laptop['Laptop']
for comment in allcomment:
    rate_word(comment,dictionary_laptop)
```

```python
print(len(dictionary_laptop))
```

```python
mydict = dictionary_laptop.copy()
for word in mydict:
    if len(word) <3:
        del dictionary_laptop[word]
```

```python
print(len(dictionary_laptop))
```

```python
stopwords = open('turkce-stop-words', 'r',encoding ='utf-8').read().split()
#stopwords
# remove the stop words from dictionary
for stops in stopwords:
    if stops in dictionary_laptop:
        del dictionary_laptop[stops]
print(len(dictionary_laptop))        
```

```python
count = 0
for x in dictionary_laptop:
    if dictionary_laptop[x] >1:
        print(x,"=", dictionary_laptop[x])
        count = count+1
print("Number Of Word ========> ",count)
```
