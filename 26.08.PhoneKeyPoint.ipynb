```python
def rate_word(comment,dictionary):
    splitComment = comment.split(" ")
    for eachcomment in splitComment:
        eachcomment = eachcomment.lower()
        if eachcomment in dictionary:
            x = dictionary[eachcomment]
            dictionary[eachcomment] = x+1
        else:
            dictionary[eachcomment] = 1
```

```python
dictionary_phone = {}
allcomment = data_df['comment']
for comment in allcomment:
    rate_word(comment,dictionary_phone)
```

```python
print(len(dictionary_phone))
```

```python
mydict = dictionary_phone.copy()
for word in mydict:
    if len(word) <3:
        del dictionary_phone[word]
```

```python
print(len(dictionary_phone))
```

```python
stopwords = open('turkce-stop-words', 'r',encoding ='utf-8').read().split()
#stopwords
# remove the stop words from dictionary
for stops in stopwords:
    if stops in dictionary_phone:
        del dictionary_phone[stops]
print(len(dictionary_phone))  
```

```python
count = 0
for x in dictionary_phone:
    if dictionary_phone[x] >1:
        print(x,"=", dictionary_phone[x])
        count = count+1
print("Number Of Word ========> ",count)
```

```python
count = 0
for x in dictionary_phone:
    if dictionary_phone[x] >32 and dictionary_phone[x] <130:
        print(x,"=", dictionary_phone[x])
        count = count+1
print("Number Of Word ========> ",count)
```
